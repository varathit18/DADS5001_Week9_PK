```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df_airbnb = pd.read_csv('https://raw.githubusercontent.com/varathit18/Implement-SGD-MiniBatch/refs/heads/main/listings.csv')
df_airbnb.head()
```

```python
df_airbnb.info()
```

```python
# prompt: df_airbnb scatter plot latitude and longitude
# add color based on room type

import plotly.express as px

fig = px.scatter(df_airbnb,
                 x="longitude",
                 y="latitude",
                 color="room_type",
                 hover_name="name",
                 title="Airbnb Listings by Location and Room Type")
fig.show()
```
